```python
# Chemin vers le fichier notes.txt
fichier = "C:\\Users\\meuni\\Downloads\\notes.txt"

# Initialisation d'une liste pour stocker les notes
notes = []

# Ouverture et lecture du fichier
with open(fichier, 'r') as fichiers:
    for ligne in fichiers:
        # Conversion de la ligne en float et ajout à la liste
        notes.append(float(ligne.strip()))

# Calcul de la moyenne
moyenne = sum(notes) / len(notes)

# Affichage de la moyenne avec deux décimales
print(f"Moyenne des notes : {moyenne}")
```

```python
# Chemin vers le fichier notes et le nouveau fichier notes2
fichier = "C:\\Users\\meuni\\Downloads\\notes.txt"
fichier2 = "C:\\Users\\meuni\\Downloads\\notes2.txt"

# Initialisation d'une liste pour stocker les notes
notes = []

# Lecture des notes du fichier original
with open(fichier, 'r') as fichiers:
    for ligne in fichiers:
        notes.append(float(ligne.strip()))

# Écriture dans le nouveau fichier
with open(fichier2, 'w') as fichiers2:
    for note in notes:
        # Détermination de l'admission ou du recalage
        resultat = "admis" if note >= 10 else "recalé"
        # Écriture de la note et du résultat dans le nouveau fichier
        fichiers2.write(f"{note:.1f} {resultat}\n")
```
