```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
from sklearn.model_selection import GridSearchCV

# Définition des hyperparamètres à tester
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10],  
    'min_samples_split': [2, 5, 10]  
}

# Création du classificateur d'arbre de décision
clf = tree.DecisionTreeClassifier()

# Création de l'objet GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=5)  # cv = 5 pour une validation croisée à 5 plis

# Entraînement de GridSearchCV sur les données
grid_search.fit(X, y)

# Affichage des hyperparamètres optimaux
print("Hyperparamètres optimaux:", grid_search.best_params_)

# Affichage de la meilleure précision obtenue
print("Meilleure précision:", grid_search.best_score_)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree

# Charger les données iris
iris = load_iris()
X, y = iris.data, iris.target

# Créer un classificateur d'arbre de décision
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

# Afficher l'arbre de décision
plt.figure(figsize=(15, 10))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
iris = load_iris()
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=2)
decision_tree = decision_tree.fit(iris.data, iris.target)
r = export_text(decision_tree, feature_names=iris['feature_names'])
print(r)
```
